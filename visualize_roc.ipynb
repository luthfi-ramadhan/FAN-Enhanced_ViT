```python
import numpy as np
import h5py
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
```

# ROC 4CH

```python
h5f = h5py.File('prediction_proposed_4ch.h5', 'r')
proposed_label = h5f['y_test'][:]
proposed_prob = h5f['y_prob'][:]
proposed_pred = h5f['y_pred'][:]
h5f.close()

h5f = h5py.File('prediction_sota_4ch.h5', 'r')
sota_label = h5f['y_test'][:]
sota_prob = h5f['y_prob'][:]
sota_pred = h5f['y_pred'][:]
h5f.close()

h5f = h5py.File('prediction_3dcnn_4ch.h5', 'r')
resnet_label = h5f['y_test'][:]
resnet_prob = h5f['y_prob'][:]
resnet_pred = h5f['y_pred'][:]
h5f.close()

h5f = h5py.File('prediction_LIFES_4ch.h5', 'r')
LIFES_label = h5f['y_test'][:]
LIFES_prob = h5f['y_prob'][:]
LIFES_pred = h5f['y_pred'][:]
h5f.close()

h5f = h5py.File('prediction_MLR_4ch.h5', 'r')
MLR_label = h5f['y_test'][:]
MLR_prob = h5f['y_prob'][:]
MLR_pred = h5f['y_pred'][:]
h5f.close()

h5f = h5py.File('prediction_new_sota_4ch.h5', 'r')
new_sota_label = h5f['y_test'][:]
new_sota_prob = h5f['y_prob'][:]
new_sota_pred = h5f['y_pred'][:]
h5f.close()
```

```python
tpr_proposed, fpr_proposed, _ = roc_curve(proposed_label, proposed_prob)
tpr_MLR, fpr_MLR, _ = roc_curve(MLR_label, MLR_prob)
tpr_LIFES, fpr_LIFES, _ = roc_curve(LIFES_label, LIFES_prob)
tpr_resnet, fpr_resnet, _ = roc_curve(resnet_label, resnet_prob)
tpr_sota, fpr_sota, _ = roc_curve(sota_label, sota_prob)
tpr_new_sota, fpr_new_sota, _ = roc_curve(new_sota_label, new_sota_prob)

proposed_auc = auc(tpr_proposed, fpr_proposed)
MLR_auc = auc(tpr_MLR, fpr_MLR)
LIFES_auc = auc(tpr_LIFES, fpr_LIFES)
resnet_auc = auc(tpr_resnet, fpr_resnet)
sota_auc = auc(tpr_sota, fpr_sota)
new_sota_auc = auc(tpr_new_sota, fpr_new_sota)

plt.plot(tpr_proposed, fpr_proposed, label='Proposed (area = %0.2f)' % proposed_auc)
plt.plot(tpr_MLR, fpr_MLR, label='ViT + Transformer (area = %0.2f)' % MLR_auc)
plt.plot(tpr_LIFES, fpr_LIFES, label='LIFES (area = %0.2f)' % LIFES_auc)
plt.plot(tpr_resnet, fpr_resnet, label='3DResNet (area = %0.2f)' % resnet_auc)
plt.plot(tpr_sota, fpr_sota, label='Vit L + BiLSTM (area = %0.2f)' % sota_auc)
plt.plot(tpr_new_sota, fpr_new_sota, label='ViViEchoformer (area = %0.2f)' % new_sota_auc)




#plt.plot(ns_fpr, ns_tpr, linestyle='--', label='Random (area = %0.2f)' % auc_random)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.title('A4C Receiver Operating Characteristic Curve')
plt.show()
```
