```python
import pickle
import numpy as np
import h5py
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
```

# 4CH

```python
h5f = h5py.File(r'ablation_fan\chosen_predict\4ch\prediction_ablation_fan_4ch.h5', 'r')
ablation_fan_label = h5f['y_test'][:]
ablation_fan_prob = h5f['y_prob'][:]
ablation_fan_pred = h5f['y_pred'][:]
h5f.close()

h5f = h5py.File(r'tune_fan_1per10\chosen_predict\4ch\prediction_tune_fan_1per10_4ch.h5', 'r')
tune_fan_1per10_label = h5f['y_test'][:]
tune_fan_1per10_prob = h5f['y_prob'][:]
tune_fan_1per10_pred = h5f['y_pred'][:]
h5f.close()

h5f = h5py.File(r'tune_fan_1per8\chosen_predict\4ch\prediction_tune_fan_1per8_4ch.h5', 'r')
tune_fan_1per8_label = h5f['y_test'][:]
tune_fan_1per8_prob = h5f['y_prob'][:]
tune_fan_1per8_pred = h5f['y_pred'][:]
h5f.close()

h5f = h5py.File(r'tune_fan_1per6\chosen_predict\4ch\prediction_tune_fan_1per6_4ch.h5', 'r')
tune_fan_1per6_label = h5f['y_test'][:]
tune_fan_1per6_prob = h5f['y_prob'][:]
tune_fan_1per6_pred = h5f['y_pred'][:]
h5f.close()

h5f = h5py.File(r'prediction_proposed_4ch.h5', 'r')
tune_fan_1per4_label = h5f['y_test'][:]
tune_fan_1per4_prob = h5f['y_prob'][:]
tune_fan_1per4_pred = h5f['y_pred'][:]
h5f.close()

h5f = h5py.File(r'tune_fan_1per2\chosen_predict\4ch\prediction_tune_fan_1per2_4ch.h5', 'r')
tune_fan_1per2_label = h5f['y_test'][:]
tune_fan_1per2_prob = h5f['y_prob'][:]
tune_fan_1per2_pred = h5f['y_pred'][:]
h5f.close()
```

```python
tpr_ablation_fan, fpr_ablation_fan, _ = roc_curve(ablation_fan_label, ablation_fan_prob)
tpr_tune_fan_1per10, fpr_tune_fan_1per10, _ = roc_curve(tune_fan_1per10_label, tune_fan_1per10_prob)
tpr_tune_fan_1per8, fpr_tune_fan_1per8, _ = roc_curve(tune_fan_1per8_label, tune_fan_1per8_prob)
tpr_tune_fan_1per6, fpr_tune_fan_1per6, _ = roc_curve(tune_fan_1per6_label, tune_fan_1per6_prob)
tpr_tune_fan_1per4, fpr_tune_fan_1per4, _ = roc_curve(tune_fan_1per4_label, tune_fan_1per4_prob)
tpr_tune_fan_1per2, fpr_tune_fan_1per2, _ = roc_curve(tune_fan_1per2_label, tune_fan_1per2_prob)

ablation_fan_auc = auc(tpr_ablation_fan, fpr_ablation_fan)
tune_fan_1per10_auc = auc(tpr_tune_fan_1per10, fpr_tune_fan_1per10)
tune_fan_1per8_auc = auc(tpr_tune_fan_1per8, fpr_tune_fan_1per8)
tune_fan_1per6_auc = auc(tpr_tune_fan_1per6, fpr_tune_fan_1per6)
tune_fan_1per4_auc = auc(tpr_tune_fan_1per4, fpr_tune_fan_1per4)
tune_fan_1per2_auc = auc(tpr_tune_fan_1per2, fpr_tune_fan_1per2)


plt.plot(tpr_ablation_fan, fpr_ablation_fan, label='dp=0 (area = %0.2f)' % ablation_fan_auc)
plt.plot(tpr_tune_fan_1per10, fpr_tune_fan_1per10, label='dp=1/10 (area = %0.2f)' % tune_fan_1per10_auc)
plt.plot(tpr_tune_fan_1per8, fpr_tune_fan_1per8, label='dp=1/8 (area = %0.2f)' % tune_fan_1per8_auc)
plt.plot(tpr_tune_fan_1per6, fpr_tune_fan_1per6, label='dp=1/6 (area = %0.2f)' % tune_fan_1per6_auc)
plt.plot(tpr_tune_fan_1per4, fpr_tune_fan_1per4, label='dp=1/4 (area = %0.2f)' % tune_fan_1per4_auc)
plt.plot(tpr_tune_fan_1per2, fpr_tune_fan_1per2, label='dp=1/2 (area = %0.2f)' % tune_fan_1per2_auc)


plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.title('A4C Receiver Operating Characteristic Curve')
plt.show()
```
